```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leitura dos dados
df = pd.read_csv('../data/simulation/random_strategy_sim.csv', nrows=76)
```

```python
# Análise Descritiva
df.describe()
```

```python
# Relacionamento entre Variáveis
df.corr()
```

```python
# Relacionamento entre Variáveis - Heatmap
corr_matrix = df.corr()

# Configuração do heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Matriz de Correlação entre Variáveis')
plt.show()
```

```python
# Comparação de Rounds
df.groupby('round').mean()
```

```python
# Comparação de Rounds
round_comparison = df.groupby('round').mean().round(2)  # Arredondar para 2 casas decimais

# Renomear as colunas
round_comparison.columns = ['trees', 'sim_time', 'mse1', 'pearson1', 'mse2', 'pearson2', 'mse3', 'pearson3', 'mse4', 'pearson4']

# Criar uma tabela para visualização com ajustes de fonte e tamanho
fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('off')  # Desativa os eixos para uma aparência mais limpa

# Ajustes de fonte e tamanho
font_size = 10
table = ax.table(cellText=round_comparison.values, colLabels=round_comparison.columns, loc='center', cellLoc='center', fontsize=font_size)
table.auto_set_font_size(False)
table.set_fontsize(font_size)

# Salvar a tabela como uma imagem
plt.savefig('random_strategy_group_by_round.png', bbox_inches='tight')
plt.show()
```

```python
# Análise de Tendências Temporais com uma métrica específica (por exemplo, mse1)
df.plot(x='sim_time', y='mse1', kind='scatter')
plt.title('MSE1 em função do tempo de simulação')
plt.xlabel('Tempo de Simulação')
plt.ylabel('MSE1')
plt.show()
```

```python
len(df[['mse1', 'mse2', 'mse3', 'mse4']])
```

```python
# Gráfico de dispersão para sim_time em relação a mse1, mse2, mse3 e mse4
plt.figure(figsize=(10, 6))

# Loop através das colunas mse1, mse2, mse3, mse4
for column in ['mse1', 'mse2', 'mse3', 'mse4']:
    plt.scatter(df['sim_time'], df[column], alpha=0.5)

plt.title('Erro Quadrático Médio em relação ao Tempo')
plt.xlabel('sim_time')
plt.ylabel('mse1, mse2, mse3, mse4')
plt.grid(True)
plt.legend(['mse1', 'mse2', 'mse3', 'mse4'])
plt.show()
```

```python
# Gráfico de dispersão para sim_time em relação a mse1, mse2, mse3 e mse4
plt.figure(figsize=(10, 6))

# Loop através das colunas mse1, mse2, mse3, mse4
for column in ['mse1', 'mse2', 'mse3', 'mse4']:
    plt.scatter(df['trees_by_client'], df[column], alpha=0.5)

plt.title('Erro Quadrático Médio em relação ao Número de Árvores por Cliente')
plt.xlabel('Número de Árvores por Cliente')
plt.ylabel('mse1, mse2, mse3, mse4')
plt.grid(True)
plt.legend(['mse1', 'mse2', 'mse3', 'mse4'])
plt.show()
```

```python
# Comparação entre Pearson_corr e número de árvores
df.plot(x='trees_by_client', y='pearson_corr1', kind='scatter')
plt.title('Pearson Corr1 em relação ao número de árvores por cliente')
plt.xlabel('Número de Árvores por Cliente')
plt.ylabel('Pearson Corr1')
plt.show()
```

```python
# Gráfico de dispersão para trees_by_client em relação a pearson_corr1, pearson_corr2, pearson_corr3 e pearson_corr4
plt.figure(figsize=(10, 6))

# Loop através das colunas pearson_corr1, pearson_corr2, pearson_corr3, pearson_corr4
for column in ['pearson_corr1', 'pearson_corr2', 'pearson_corr3', 'pearson_corr4']:
    plt.scatter(df['trees_by_client'], df[column], alpha=0.5)

plt.title('Correlação de Pearson em relação ao Número de Árvores por Cliente')
plt.xlabel('trees_by_client')
plt.ylabel('pearson_corr1, pearson_corr2, pearson_corr3, pearson_corr4')
plt.grid(True)
plt.legend(['pearson_corr1', 'pearson_corr2', 'pearson_corr3', 'pearson_corr4'])
plt.show()
```

```python
# Plotar gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['sim_time'], alpha=0.5)
plt.title('Relação entre Tempo de Simulação e Iteração da Simulação')
plt.xlabel('iteration')
plt.ylabel('sim_time')
plt.grid(True)
plt.show()
```
